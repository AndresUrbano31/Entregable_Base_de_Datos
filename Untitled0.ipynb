```python
#bibliotecas para procesar la informacion
import numpy as np #arreglos, media,mediana
import scipy.stats as stats #para sacar la moda
import matplotlib.pyplot as plt # graficar

# Datos de ejemplo
data = np.array([2, 3, 5, 7, 7, 8, 9, 10, 10, 10, 11, 13, 13, 14, 16])

# Cálculos
media = np.mean(data) # media= promedio
mediana = np.median(data) #mediana = diferencia entre unidades - la diferencia entre media y mediana da la desviacion estandar
moda = stats.mode(data, keepdims=True).mode[0] #moda que dato mas se repite
#data frame = se usa pandas se trabaja con diccionarios
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

# Visualización
plt.hist(data, bins=range(2, 18), alpha=0.7, color='blue', edgecolor='black') # histograma
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media = {media}') # grafica las lineas de media,mediana,moda
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=1, label=f'Mediana = {mediana}')
plt.axvline(moda, color='orange', linestyle='dashed', linewidth=1, label=f'Moda = {moda}')
plt.legend()
plt.title('Histograma con Media, Mediana y Moda')
plt.xlabel('Valor') #eje x
plt.ylabel('Frecuencia') #eje y
plt.show() #muestra la grafica
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Cálculos
varianza = np.var(data, ddof=1) # mirar la diferencia entre cada dato y la media
desviacion_estandar = np.std(data, ddof=1) # cuando hay una diferencia amplia se debe revizar si hay valore no-null

print(f"Varianza: {varianza}")
print(f"Desviación Estándar: {desviacion_estandar}")

# Visualización
plt.hist(data, bins=range(2, 18), alpha=0.7, color='blue', edgecolor='black') #histograma solo valores positivos para valores negativos se usa  diagrama de flujo - diagrama con cuartiles
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media = {media}')
plt.fill_betweenx([0, 5], media-desviacion_estandar, media+desviacion_estandar, color='orange', alpha=0.3, label=f'±1 Desviación Estándar')
plt.legend()
plt.title('Histograma con Desviación Estándar')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
data = np.array([2, 3, 5, 7, 7, 8, 9, 10, 10, 10, 11, 13, 13, 14, 16])

# Cálculo de Cuartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # También es la mediana
Q3 = np.percentile(data, 75)

# Cálculo de Percentiles
percentil_10 = np.percentile(data, 10)
percentil_90 = np.percentile(data, 90)

# Mostrar resultados de cuartiles y percentiles
print(f"Primer Cuartil (Q1): {Q1}") #25%
print(f"Mediana (Q2): {Q2}") # 50%
print(f"Tercer Cuartil (Q3): {Q3}") #75%
print(f"Percentil 10: {percentil_10}")
print(f"Percentil 90: {percentil_90}")

# Crear el diagrama de cajas
plt.boxplot(data, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'))

# Añadir líneas para los percentiles 10 y 90
plt.axvline(percentil_10, color='purple', linestyle='dotted', linewidth=1, label=f'Percentil 10 = {percentil_10}')
plt.axvline(percentil_90, color='brown', linestyle='dotted', linewidth=1, label=f'Percentil 90 = {percentil_90}')

# Etiquetas y título
plt.title('Diagrama de Cajas con Cuartiles y Percentiles')
plt.xlabel('Valor')
plt.legend()

# Mostrar la gráfica
plt.show()

#en este diagrama se busca que los datos a encontrar sean pocos
```
